```python
tpf_file = lk.search_targetpixelfile('TIC 467179528', mission="TESS", sector=76).download(quality_bitmask='default')
```

```python
%matplotlib inline
tpf_file.plot();
```

LIGHT CURVE AND FOLDING

```python
lc = tpf_file.to_lightcurve(aperture_mask=tpf_file.pipeline_mask)
lc
```

```python
%matplotlib inline
lc.plot();
```

```python
flat_lc = lc.flatten(window_length=401)
flat_lc.plot();
```

```python
folded_lc = flat_lc.fold(period=	10.895,  epoch_time=2459649.7512990)
folded_lc.plot();
```
